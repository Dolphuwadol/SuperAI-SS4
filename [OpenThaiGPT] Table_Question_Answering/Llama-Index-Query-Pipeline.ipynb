```python
import pandas as pd

df = pd.read_csv("TBL4-Online-Shopping-Dataset.csv")
df
```

```python
instruction_str = (
    "1. Convert the query to executable Python code using Pandas.\n"
    "2. The final line of code should be a Python expression that can be called with the `eval()` function.\n"
    "3. The code should represent a solution to the query.\n"
    "4. PRINT ONLY THE EXPRESSION.\n"
    "5. Do not quote the expression.\n"
    "6. Consider that, when asked about total number, you need to count for UNIQUE value ID by simply use .unique() pandas function\n"
    "7. Don't use df.duplicated no matter what."

    """Consider these fields:
CustomerID (Numeric)
Description: Unique identifier for each customer.

Gender (Categorical)
Description: Gender of the customer (e.g., M, F).

Location (Text)
Description: Location or address information of the customer.

Tenure_Months (Numeric)
Description: Number of months the customer has been associated with the platform.

Transaction_ID (Numeric)
Description: Unique identifier for each transaction.

Transaction_Date (Date)
Description: Date of the transaction.

Product_SKU (Text)
Description: Stock Keeping Unit (SKU) identifier for the product.

Product_Description (Text)
Description: Description of the product.

Product_Category (Categorical)
Description: Category to which the product belongs.

Quantity (Numeric)
Description: Quantity of the product purchased in the transaction.

Avg_Price (Numeric)
Description: Average price of the product.

Total_Price (Numeric)
Description: Total price of the product exclude delivery charges.

Delivery_Charges (Numeric)
Description: Charges associated with the delivery of the product.

Date (Date)
Description: Date of the transaction (potentially redundant with Transaction_Date).

Month (Numeric)
Description: Month of the transaction."""
)

pandas_prompt_str = (
    "You are working with a pandas dataframe in Python.\n"
    "The name of the dataframe is `df`.\n"
    "This is the result of `print(df.head())`:\n"
    "{df_str}\n\n"
    "Follow these instructions:\n"
    "{instruction_str}\n"
    "Query: {query_str}\n\n"
    "Expression:"
)

response_synthesis_prompt_str = (
    "Given an input question, synthesize a response from the query results.\n"
    "Query: {query_str}\n\n"
    "Please follow this instruction carefully\n"
    "If the answer is a number, simply answer the number, DO NOT add any other text"
    "Example: Do this : 100\nDo NOT DO this: The number of customer is 100."
    "If there're more than 2 decimal points, only print the first two."
    "Pandas Instructions (optional):\n{pandas_instructions}\n\n"
    "Pandas Output: {pandas_output}\n\n"
    "Response: "
)

pandas_prompt = PromptTemplate(pandas_prompt_str).partial_format(
    instruction_str=instruction_str, df_str=df.head(5)
)
pandas_output_parser = PandasInstructionParser(df)
response_synthesis_prompt = PromptTemplate(response_synthesis_prompt_str)
```

```python
# specify how to quantize the model
quantization_config = BitsAndBytesConfig(
        load_in_8bit=True,
        bnb_8bit_quant_type="nf8",
        bnb_8bit_compute_dtype=torch.float16,
)

model = AutoModelForCausalLM.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    token=<hf token>,
    device_map="auto",
    quantization_config=quantization_config,
)

tokenizer = AutoTokenizer.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    token=<hf token>
)
llm = HuggingFaceLLM(
    model=model,
    tokenizer=tokenizer
)
```

## Build Query Pipeline

Looks like this:
input query_str -> pandas_prompt -> llm1 -> pandas_output_parser -> response_synthesis_prompt -> llm2

Additional connections to response_synthesis_prompt: llm1 -> pandas_instructions, and pandas_output_parser -> pandas_output.

```python
qp = QueryPipeline(
    modules={
        "input": InputComponent(),
        "pandas_prompt": pandas_prompt,
        "llm1": llm,
        "pandas_output_parser": pandas_output_parser,
        "response_synthesis_prompt": response_synthesis_prompt,
        "llm2": llm,
    },
    verbose=True,
)
qp.add_chain(["input", "pandas_prompt", "llm1", "pandas_output_parser"])
qp.add_links(
    [
        Link("input", "response_synthesis_prompt", dest_key="query_str"),
        Link(
            "llm1", "response_synthesis_prompt", dest_key="pandas_instructions"
        ),
        Link(
            "pandas_output_parser",
            "response_synthesis_prompt",
            dest_key="pandas_output",
        ),
    ]
)
# add link from response synthesis prompt to llm2
qp.add_link("response_synthesis_prompt", "llm2")
```
